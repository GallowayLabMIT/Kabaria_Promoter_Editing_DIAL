## Calculate and Plot Fold Change, ZF37

### This is also the green figures plots of gmean and fold change

```python
ZFs = pd.array(['noZF','ZF37'])
s_ZF37foldchange = s[ s['ZF'].isin(ZFs)]
s_ZF37foldchange = s_ZF37foldchange[ s_ZF37foldchange['reporter']=='pTET021_203bp']
bios = pd.unique(s_ZF37foldchange['bioreplicate'] )

map_noGIB = {'test':'test'}

for biorep in bios:
    # Normalize all to the noCre ZF37 conditions
    mGL_noCre_ZF37 =  s_ZF37foldchange[ s_ZF37foldchange['ZF']=='ZF37']
    mGL_noCre_ZF37 =  mGL_noCre_ZF37[ mGL_noCre_ZF37['bioreplicate']==biorep]
    mGL_noCre_ZF37 =  mGL_noCre_ZF37[ mGL_noCre_ZF37['GIB']=='0xGIB']
    mGL_noCre_ZF37= mGL_noCre_ZF37['mGL-A_gmean'].iloc[0]
    map_noGIB[biorep] = mGL_noCre_ZF37

s_ZF37foldchange['mGL_noGIB'] = s_ZF37foldchange['bioreplicate'].map(map_noGIB)
s_ZF37foldchange['Fold Change'] = s_ZF37foldchange['mGL-A_gmean'] / s_ZF37foldchange['mGL_noGIB']


rept = 'pTET021_203bp'
x = 'cond'; x_rot = 45
xlabel = ''
fontsize = 14
xsize =4/3*2  ; ysize = 3
fontsize = 14
hue = None;  hueorder = None; 
yscales = pd.array(['log'])
ylim = [0,5]
pairs = [ 
    [rept + '.ZF37.0xGIB',rept +  '.ZF37.1xGIB'], 
        [rept + '.noZF.0xGIB',rept +  '.noZF.1xGIB'] ]
palette = {
        rept + '.noZF.0xGIB': 'grey',
        rept + '.noZF.1xGIB': 'black',
        rept + '.ZF37.0xGIB': '#C3DB9A',
        rept + '.ZF37.1xGIB': '#69B645'}
order =  palette
data_now = s_ZF37foldchange
y = 'Fold Change'
ylabel = 'Output Fold Change (mGL)'
yscale = 'linear'
plottitle = rept + ' ' + y + 'mGL only'
g = plt.figure(figsize=(xsize,ysize))
plt.axhline(y=1, ls='--', color='#C3DB9A')
g = summary_plot( x, y, hue, data=data_now, dodge=True, palette = palette, order=order, 
                hue_order = hueorder, plottitle=plottitle,yscale=yscale,x_rot = x_rot,
                ylim =ylim, fontsize=fontsize,ylabel=ylabel,xlabel=xlabel,xticklabels=False,
                pairs=pairs )
g.figure.savefig(figure_folder + plottitle + '.svg',dpi=300,bbox_inches='tight')

y = 'mGL-A_gmean'
ylabel = 'Output gmean (mGL)'
yscale = 'log'
ylim = [3e1, 1e4]
plottitle = rept + ' ' + y + 'mGL only'
g = plt.figure(figsize=(xsize,ysize))
g = summary_plot( x, y, hue, data=data_now, dodge=True, palette = palette, order=order, 
                hue_order = hueorder, plottitle=plottitle,yscale=yscale,x_rot = x_rot,
                ylim =ylim, fontsize=fontsize, ylabel=ylabel, xticklabels=False, xlabel=xlabel,
                pairs=pairs )
g.figure.savefig(figure_folder + plottitle + '.svg',dpi=300,bbox_inches='tight')
```
